Importing Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading Dataset

```python
df=pd.DataFrame()
df = pd.read_csv('http://raw.githubusercontent.com/mwaskom/seaborn-data/master/car_crashes.csv?utm_source=chatgpt.com')
df.head()
```

Basic Info

Analysis

```python
#Total Accidents Distribution

import matplotlib.pyplot as plt
df['total'].plot(kind='hist')
plt.title('Total Accidents Distribution')
plt.xlabel('Total Accidents')
plt.show()

# Speeding vs Total Accidents 🔥

plt.scatter(df['speeding'], df['total'])
plt.xlabel('Speeding')
plt.ylabel('Total Accidents')
plt.title('Speeding vs Total Accidents')
plt.show()

# Alcohol vs Total Accidents 🔥

plt.scatter(df['alcohol'], df['total'])
plt.xlabel('Alcohol')
plt.ylabel('Total Accidents')
plt.title('Alcohol vs Total Accidents')
plt.show()

# Insurance Losses Distribution

df['ins_losses'].plot(kind='hist')
plt.title('Insurance Losses Distribution')
plt.show()
```

Conclusions :


The analysis of accident data shows that factors such as speeding and alcohol consumption have a strong relationship with total accidents. Higher values of these factors correspond to increased accident rates. Additionally, insurance losses are higher in regions with more accidents, indicating economic impact.
